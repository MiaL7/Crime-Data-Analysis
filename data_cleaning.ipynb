```python
import pandas as pd
import numpy as np
import zipfile
import os
from datetime import datetime
import pickle
```

```python
with zipfile.ZipFile("data.zip", "r") as z:
    with z.open("Crime_Data_from_2020_to_Present.csv") as csv_file:
        data = pd.read_csv(csv_file)

data.head()
```

```python
data["Date Rptd"] = data["Date Rptd"].str.replace(" 12:00:00 AM", "")
data["DATE OCC"] = data["DATE OCC"].str.replace(" 12:00:00 AM", "")
data["Date Rptd"] = pd.to_datetime(data["Date Rptd"], format = '%m/%d/%Y')
data["DATE OCC"] = pd.to_datetime(data["DATE OCC"], format = '%m/%d/%Y')
```

```python
data.head()
```

```python
df = data.set_index("DATE OCC").sort_index()
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(axis = 1, columns=['Date Rptd',
                                'Part 1-2',
                                'Mocodes', 
                                'Weapon Used Cd', 
                                'Weapon Desc', 
                                'Crm Cd 1',
                                'Crm Cd 2', 
                                'Crm Cd 3', 
                                'Crm Cd 4', 
                                'Cross Street'])
```

```python
df.isna().sum()
```

```python
df['Premis Cd'].fillna(0, inplace = True)
```

```python
df['Premis Desc'].fillna('UNKNOWN', inplace = True)
```

```python
df['Vict Sex'].value_counts()
```

```python
df['Vict Sex'].replace(to_replace=['H', '-'], 
                       value = 'X', inplace = True)
df['Vict Sex'].fillna('X', inplace= True)
```

```python
df['Vict Age'].value_counts()
```

```python
df['Vict Age'] = np.where((df['Vict Age'] < 0), 0, df['Vict Age'])
```

```python
df['Vict Age'].value_counts()
```

```python
df.isna().sum()
```

```python
df['Vict Descent'].value_counts()
```

```python
df['Vict Descent'].replace(to_replace = ['-'], 
                           value = 'X', inplace = True)
```

```python
df['Vict Descent'].fillna('X', inplace = True)
```

```python
df['Vict Descent'].value_counts()
```
